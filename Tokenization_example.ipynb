```python
corpus = '''Hello I am Shivansh, I am a AI Researcher.
I am a 3rd year undergrad Student.
'''
```

```python
print(corpus)
```

```python
## Tokenization
## Sentence ->paragraphs
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```
